Step 2: Scatterplot graphs


```python
#Loading the required modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Separating the data

brainFrame = pd.read_csv(brainFile, delim_whitespace=True)
menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

```python
#Plotting the graphs

menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
#Plotting the Female gender data

womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.show()
%matplotlib inline
```

Analysis: Since we created a separate dataset for the Genders Male and Female, the function only creates a scatter plot that denotes the MRI Count of Male Patient. This is to ensure that the data presented is only for a specific demographic of people. This could also be similarly created for the Female demographic.

# Part 3: Calculating Correlation with Python

```python
#Calculate the correlation against brainFrame

brainFrame.corr(method='pearson')
```

Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filled with 1s? Is that a coincidence? Explain.


*   The repetitive pattern of diagonal 1s is not a coincidence, as this simply means that the computed value of the variable is perfectly correlated with itself.


Still looking at the correlation table above, notice that the values are mirrored; values below the 1 diagonal have a mirrored counterpart above the 1 diagonal. Is that a coincidence? Explain.

*   This means that the matrix is symmetrical.

# Statistical Analysis of Age and the Diabetes Patients Concentration

```python
#Loading the diabetes dataset

diabetesFile = 'diabetes.csv'
diabetesFrame = pd.read_csv(diabetesFile)
```

```python
#Separating the datasets by age group

AdultsDf = diabetesFrame[(diabetesFrame.Age <= 45 )]
Adults1f = diabetesFrame[(diabetesFrame.Age > 45)]
```

```python
#Adults 45 and Above

Acorr = AdultsDf.corr()
sns.heatmap(Acorr)
```

```python
#Adults below 45

Ycorr = Adults1f.corr()
sns.heatmap(Ycorr)
```

# Analysis

After categorizing the dataset by a certain range of age, it is shown in the heatmap that people above the age of 45 have a higher level of glucose as compared to ages below 45, as such it could be seen from the age and outcome variables that there is a higher concentration of above 45 developing diabetes.

Although the relationship is rather weak, as scrolling through the database it could be seen that in some cases, people below 45 had developed diabetes, the heatmap cannot be a definitive representation that ages above 45 do in fact develop diabetes, but only in the sense that they are in a higher risk.

The age and Diabetes Pedigree (risk) also presents a weak relationship, as analyzing the table shows that there are some cases where the risk factor is low, and they still developed diabetes.
